#### 2.2 Calculate Returns for all stocks with inference

```python
stock_returns = np.log(DF).diff(1)
```

```python
stock_returns.head()
```

```python
stock_returns.describe()
```

#### 2.3 Calculate Stock Means and Standard Deviation for all stocks with inference

```python
stock_mean = stock_returns.mean()
stock_mean.sort_values(ascending=False)
```

```python
stock_std = stock_returns.std()
stock_std.sort_values(ascending=False)
```

Shree Cement has the highest returns while Idea Vodafone has the lowest. The market risk (volatility) associated is the highest for Jet Airways while Infosys is the least risky option to invest in.

#### 2.4 Draw a plot of Stock Means vs Standard Deviation and state your inference

```python
Mean = stock_mean.to_frame(name='Stock Mean')
STD = stock_std.to_frame(name='Stock Standard Deviation')
```

```python
Mean_STD = pd.merge(Mean, STD, right_index = True, left_index = True)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(Mean_STD['Stock Mean'],Mean_STD['Stock Standard Deviation'], hue=Mean_STD.index, legend="full")
plt.title('Mean Vs Standard Deviation')
plt.tight_layout()
plt.show()
```

1. stocks higher up on the top far left indicate higher volatility and low returns.
2. stocks at the lower bottom right indicate low volatility and higher returns.

#### 2.5 Conclusion and Recommendations

Stocks that have lower mean (returns) and higher standard deviation (volatility) do not really play a vital role in a portfolio that contains other stocks with higher returns and low risk. Hence, we need to focus on:
1. stocks that have the highest returns with lowest risk
2. stocks that have the lowest risk and highest returns

therefore, based on the data set, stocks thats have highest returns are Shree Cement, followed by Infosys and Axis Bank.
in terms of lowest risk we have Infosys, followed by Mahindra & Mahindra and Indian Hotel.

Stock means versus Standard Deviation graph gives us insights on risk and returns. High volatility stocks may give higher returns but it may not be a good investment for long-term. Similarly, low volatile stocks may not give high returns in short-term, but might give high returns for a long term investment. Therefore, based on the type of investment one's looking for, the above plot can be used to make inferences to select the best stock to invest in.
